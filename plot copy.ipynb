```python
import pandas as pd

# read results.csv

df = pd.read_csv('results_new.csv')
# drop first two columns
df = df.drop(df.columns[[0, 1]], axis=1)
# remove rows with any -1
df = df[df.ge(0).all(axis=1)]
# remove rows with any 0


# df =df.head(1)
# split into two dataframes, with odd and even columns
df_rate = df.iloc[:, ::2]
df_error = df.iloc[:, 1::2]
# df_rate = df1.iloc[:,::2]
# df_error = df2.iloc[:,::2]
# rename columns
df_rate.columns = [i.split('_')[-1] for i in df_rate.columns]
df_error.columns = [i.split('_')[-1] for i in df_error.columns]
```

```python
df_rate_mean = df_rate.mean(axis=0)
df_error_mean = df_error.mean(axis=0)

# plot a line plot of the error and rate vs error bound using plotly

import plotly.graph_objects as go

# Create a figure
fig = go.Figure()

# Add the first trace (Series 1) on the primary y-axis
fig.add_trace(go.Scatter(x=df_rate_mean.index, y=df_rate_mean.values, mode='lines+markers', name='Rate',
                         line=dict(color='blue', width=2), marker=dict(size=8)))

# Add the second trace (Series 2) on the secondary y-axis
fig.add_trace(go.Scatter(x=df_error_mean.index, y=df_error_mean.values, mode='lines+markers', name='Error',
                         line=dict(color='red', width=2), marker=dict(size=8), yaxis='y2'))

# Update layout to include two y-axes
fig.update_layout(
    title='Error and Compression rate vs Error Bound',
    xaxis_title='Error Bounds',
    # yaxis_title='Rate',  # Title for primary y-axis
    # yaxis2_title='Errors',  # Title for secondary y-axis
    yaxis=dict(title='Rate'),  # Primary y-axis on the left
    yaxis2=dict(type = "log",overlaying='y', side='right', title= "Errors"),  # Secondary y-axis on the right
    template='plotly_white',
    showlegend=True
)

# Show the plot
fig.show()
```
